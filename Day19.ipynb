```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
```

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder \
    .appName("Read CSV example") \
    .getOrCreate()
```

```python
data = [
    ("U1", 85),
    ("U2", 72),
    ("U3", 40)
]
```

```python
df=spark.createDataFrame(data,["user_id","score"])
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
```

```python
def classify(score):
    if score >= 80:
        return "High"
    elif score >= 50:
        return "Medium"
    else:
        return "Low"
performance_udf = udf(classify, StringType())
df=df.withColumn("performance", performance_udf(col("score")))
df.show()
```

```python
from pyspark.sql.functions import when,col
df.withColumn("performance",when(col("score")>=80,"High").when(col("score")>=50,"Medium").otherwise("Low")).show()
```

```python
df.orderBy("performance", df.score.desc()).show()
```

```python
data_a=[
    ("U1","Python"),
    ("U2","Java"),
    ("U3","Spark")
]
df_a=spark.createDataFrame(data_a,["user_id","course"])
```

```python
data_b=[
    ("U2","Java"),
    ("U3","Spark"),
    ("U4","Python")
]
df_b=spark.createDataFrame(data_b,["user_id","course"])
```

```python
df_a.union(df_b).show()
```
